## Reshaping, Min, Max, Sort

## Indexing and Slicing

```python
# 1 ~ 15 series(array)
arr = np.arange(1, 16)
print(arr)

# Indexing and Slicing
print(arr[5])
print(arr[2:5])
print(arr[:6])
print(arr[-1::-1])
print(arr[-1:-6:-2])

# subset
arr = np.arange(0, 31, 3)
print(arr)
print(arr[[3, 7, 2]])
```

```python
# Matrix of random 2-digit numbers
mat = np.array(ri(10, 100, 15)).reshape(3, 5)
print(mat)

# Double Bracket Indexing
print(mat[1][3])

# Entire Row
print(mat[2])

# Entire Column
print(mat[:,3])

# Subsetting sub-matrices
print(mat[1:2, 3:5])
print(mat[0:2, [1,3]])
```

### Conditional Subsetting

```python
# Conditional Subsetting
print(mat)
print(mat[mat>40])
```

### Slicing Keeps the Original Reference

```python
mat = np.array([[11, 12, 13], [21, 22, 23], [31, 32, 33]])
print(mat)

# Slice Matrix
sliced_mat = mat[:2, :2]
print(sliced_mat)

# Change the sliced matrix
sliced_mat[0, 0] = 1000
print(sliced_mat)

# Original Matrix -> Changed!!
print(mat)

# Different Way to Creat a Copy of the Sliced Matrix
mat = np.array([[11, 12, 13], [21, 22, 23], [31, 32, 33]])
print(mat)

# Slice Matrix : np.array() -> Creates New Array
sliced_mat = np.array(mat[:2, :2])
print(sliced_mat)

# Change the sliced matrix
sliced_mat[0, 0] = 10000
print(sliced_mat)

# Original Matrix : Doesn't Change!!
print(mat)
```

## Array Operations (array-array, array-scalar, universal functions)

```python
mat1 = np.array(ri(1, 10, 9)).reshape(3,3)
mat2 = np.array(ri(1, 10, 9)).reshape(3,3)
print(mat1)
print(mat2)

# Addition
print(mat1 + mat2)
# Multiplication
print(mat1 * mat2)
# Division
print(mat1 / mat2)
# Linear Combination
print(3 * mat1 - 7 * mat2)

# Addition of a scalar
print(233 + mat1)

# Exponentiation
# 1) Matrix Cubed
print(mat1 ** 3)
# 2) Sq-Root Using Pow function
print(pow(mat1, 0.5))
```

### Broadcasting

```python
start = np.zeros((4, 3))
print(start)
```

```python
# A Rank 1 ndarray with 3 values
add_rows = np.array([1, 0, 2])
print(add_rows)
```

```python
# Broadcasting -> Add to Each Row of "start"
y = start + add_rows
print(y)
```

```python
# 4 x 1 ndarray
add_cols = np.array([[0, 1, 2, 3]])
add_cols = add_cols.T # Transpose
print(add_cols)
```

```python
# Broadcasting -> Add to each Column of "start"
y = start + add_cols
print(y)
```

```python
# Scalar
add_scalar = np.array([243])
print(start + add_scalar)
```

### NumPy Mathematical Functions on Array

```python
mat1 = np.array(ri(1, 10, 9)).reshape(3, 3)
mat2 = np.array(ri(1, 10, 9)).reshape(3, 3)
print(mat1)
print(mat2)

# Sq-Root using np
print(np.sqrt(mat1))
# Exponential Power using np
print(np.exp(mat1))
# logarithm using np
print(np.log10(mat1))
# Modulo reminder
print(np.fmod(mat1, mat2))

# Combination of Functions by Showing Exponential Decay of a Sine Wave
## Exponential Decay of Sine Wave: exp(-x)*sin(x)
A = np.linspace(0, (12 * np.pi), 1001)
import matplotlib.pyplot as plt
plt.scatter(x = A, y = (100 * np.exp(-A / 10) * (np.sin(A))))
plt.show()
```

### NumPy Basic Statistics on Array

```python
mat1 = np.array(ri(1, 10, 9)).reshape(3, 3)
mat2 = np.array(ri(1, 10, 9)).reshape(3, 3)
print(mat1)
print(mat2)

# Sum of all numbers of matrix
print(np.sum(mat1))
# Sum of all numbers in columns of matrix
print(np.sum(mat1, axis=0))
# Sum of all numbers in rows of matrix
print(np.sum(mat1, axis=1))
# Product of all numbers in rows of matrix
print(np.prod(mat1, axis=1))
# Product of all numbers in columns of matrix
print(np.prod(mat1, axis=0))
# Mean of all numbers in matrix
print(np.mean(mat1))
# Standard deviation of all numbers in matrix
print(np.std(mat1))

# A matrix with larger elements
mat1 = np.array(ri(1, 100, 9)).reshape(3, 3)
print(mat1)
# Standard Deviation
print(np.std(mat1))
# Variance
print(np.var(mat1))
# Median
print(np.median(mat1))

# 10 x 5 Matrix with random numbers form 1 to 99
mat2 = np.array(ri(1, 100, 50)).reshape(10, 5)
print(mat2)
# Standard deviation along the columns in matrix
print(np.std(mat2, axis=0))

# 4 x 5 Matrix with random numbers from 1 to 99
mat1 = np.array(ri(1, 100, 20)).reshape(4, 5)
print(mat1)
# Flattened and sorted matrix (as vector)
print(np.sort(mat1.reshape(1, 20)))
# 50th percentile of all numbers in the matrix (=median)
print(np.percentile(mat1, 50))
print(np.median(mat1)) # median
# 90th percentile of all numbers in the matrix
print(np.percentile(mat1, 90))
```

### Corelation and Covariance

```python
# Correlation Coefficient Matrix

## 1)
A = ri(1, 5, 20)
B = (2 * A) + (5 * np.random.randn(20)) #B: 2 x A + (some random noise)
### scatter plot
plt.scatter(A, B)
plt.title("Scatter Plot of A vs B")
plt.show()
### correlation coefficient matrix
print(np.corrcoef(A, B))

## 2)
A = ri(1, 50, 20)
B = 100 - (2 * A) +(10 * np.random.randn(20))
plt.scatter(A, B)
plt.title("Scatter Plot of A vs B -> Larger and Negative Correlation")
plt.show()
print(np.corrcoef(A, B))
```

## Linear Algebra Operations

### Dot/Inner/Outer Products

```python
# 1)
A = np.arange(1, 10).reshape(3, 3)
B = ri(1, 10, 9).reshape(3, 3)
print(A)
print(B)

## For 2-D: dot product = matrix multiplication
print(np.dot(A, B))

# 2)
A = np.arange(1, 6)
B = ri(1, 10, 5)
print(A)
print(B)

## Inner product
print(np.inner(A, B))
## Outer Product
print(np.outer(A, B))
```

### Transpose

```python
# 1)
A = ri(1, 10, 9).reshape(3, 3)
print(A)
## transpose
print(np.transpose(A))

#2)
B = ri(1, 10, 6).reshape(3, 2)
## transpose
print(np.transpose(B))
## Matrix Multiplication
print(np.dot(B, np.transpose(B)))
```

### Trace

```python
A = ri(1, 10, 16).reshape(4, 4)
print(A)

# Matrix trace
print(np.trace(A))
# Matrix trace with offset +1
print(np.trace(A, offset=1))
# Matrix trace with offset -1
print(np.trace(A, offset=-1))
```

### SVD(Singular Value Decomposition)

```python
A = np.random.randint(1, 10, 9).reshape(3, 3)
print(A)
print("\n")

# Singular value decomposition
u, s, v = np.linalg.svd(A, compute_uv=1, full_matrices=True)
print("u:\n", u)
print("\n")
print("v:\n", v)
print("\n")
## Reconstruction of A: A = u * s * v
print("A:\n", np.dot(u, np.dot(np.diag(s), v)))
```

### QR Decomposition/Factorization

```python
A = np.random.randint(1, 10, 9).reshape(3, 3)
print(A)
print("\n")

# QR Decomposition
q, r = np.linalg.qr(A)
print("Q:", q)
print("\n")
print("R:", r)
print("\n")
## Reconstruction of A: A = Q * R
print(np.dot(q, r))
```

### Eigenvalues and Eigenvectors

```python
A = np.random.randn(9).reshape(3, 3)
print(A)
print("\n")

# Eigenvalues and Eigenvectors
w, v = np.linalg.eig(A)
print("Eigenvalues:", w)
print("\n")
print("Eigenvectors:", v)
```

### Linear Equation Solving, Matrix Inverse, Linear Least Square

```python
# Solving Linear Equations
'''
2x + 5y + z = 14
3x - 2y - z = -1
x - 3y + z = 4
'''
A = np.array([[2, 5, 1], [3, -2, -1], [1, -3, 1]])
B = np.array([14, -1, 4])
x = np.linalg.solve(A, B)

print("Solutions: ",x)
```

```python
# Linear Least Squares(Solution to a Linear Regression Problem)
x = np.arange(1, 11, 1)
y = 7*x + np.random.randn(10) - 1
print(x)
print(y)
```

```python
# stack ones to x vertically
A = np.vstack([x, np.ones(len(x))]).T
print(A)
```

```python
# Finding coefficient and intercept
m, c = np.linalg.lstsq(A, y)[0]
print("Coefficient: ", m)
print("\n")
print("Intercept: ", c)
print("\n")

# Plot the line
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(x, y,'o', label='Original Data', markersize=8)
plt.plot(x, m*x + c, 'r', label='Fitted Line')
plt.legend()
plt.show()
```

```python
# Caculating normal multiplicative matrix
A = 0.1 * np.random.randint(1, 50, 16).reshape(4, 4)
print(A)
print("\n")
print("Inverse of A:\n", np.linalg.inv(A))
```

```python
# Calculating Moore-Penrose pseudo-inverse matrix
A = 0.1 * np.random.randint(1, 56, 15).reshape(5, 3)
print(A)
print("\n")
print("Pseudo-inverse of A:\n", np.linalg.pinv(A))
print("\n")
print("Product of A and pseudo inverse:\n", np.dot(np.linalg.pinv(A), A))
```
