## 결측치

```python
# 각 값이 결측치인지 아닌지 하나하나 확인 -> 보기 힘듦
df.isnull()
```

```python
# 각 column 별로 결측치 몇 개인지 확인
df.isnull().sum()
```

```python
# 결측치 처리(채우기, 열 삭제, 행 삭제)
```

```python
# 평균 구하기(Age의 평균)
df['Age'].mean()
```

```python
# Age의 결측치에 0 들어간 series 반환
df['Age'].fillna(0) # 이러면 원래 dataframe에 영향 안줌
```

```python
# Age의 결측치를 전부 "평균"으로 채우고 그걸 실제 데이터에 반영
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Age']
```

```python
# Age의 결측치를 전부 "중앙값"으로 채우고 그걸 실제 데이터로 반영
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Age']
```

```python
# 이제 다시 결측치 있나 확인 -> Age 결측치 0
df.isnull().sum()
```

```python
# df_original에 원래 df 복사해두고 df에서 Cabin 열 삭제
# column: axis = 1
df_original = df.copy()
df = df.drop('Cabin', axis=1)
df.head()
```

```python
# df에서 2번째 행 삭제
# row: axis = 0
df = df.drop(1, axis=0)
df.head()
```

```python
# Embarked S, C, Q 각각 몇 명인지?
df['Embarked'].value_counts()
```

```python
# Embarked 열의 최빈값 구하기 -> 0번째로 많이 나온 값 구하기
df['Embarked'].mode()[0]
```

```python
# Embarked 열 최빈값으로 Embarked 열 결측치 채우기
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Embarked']
```

```python
# 결측치 확인 -> 이제 결측치 완전히 없음
df.isnull().sum()
```

## 범주형 데이터 처리

```python
df.head()
```

```python
# Sex 원핫인코딩
'''
1) 새로운 데이터프레임 만든다
2) male 열, female 열 만든다
3) male인 사람: male = 1, female = 0
   female인 사람: male = 0, female = 1
'''
df_sex = pd.get_dummies(df['Sex'])
df_sex.head()
```

```python
# df_encoding에 df랑 df_sex 합치기
df_encoding = pd.concat([df, df_sex], axis=1)
df_encoding.head(1)
```

```python
# Embarked 원핫인코딩
df_embarked = pd.get_dummies(df_encoding['Embarked'])
df_encoding = pd.concat([df_encoding, df_embarked], axis=1)
df_encoding.head(1)
```

```python
# Sex, Embarked 열 제거
df_encoding = df_encoding.drop(['Sex', 'Embarked'], axis=1)
df_encoding.head(1)
```

```python
# df_sex, df_embarked 필요없으니 삭제
del df_sex, df_embarked
```

```python
# Ticket 열에 어떤 값 몇개씩 있는지 확인 -> Ticket 종류 680가지 있음
df_encoding['Ticket'].value_counts()
```

```python
# Ticket 열의 값 종류 확인(중복되는거 한번씩만 보여줌)
df_encoding['Ticket'].unique()
```

```python
# 숫자로 처리하기 어렵거나 필요없는 행 삭제(PassengerId, Name, Ticket)
df_encoding = df_encoding.drop('PassengerId', axis=1)
df_encoding = df_encoding.drop('Name', axis=1)
df_encoding = df_encoding.drop('Ticket', axis=1)
df_encoding.head()
```
